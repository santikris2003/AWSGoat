```python
from google import genai

client = genai.Client(api_key=userdata.get("GEMINI_API_KEY"))

# List all models
models = client.models.list()

print("Available models:")
for model in models:
    print(f"\nModel: {model.name}")
    # Print all attributes to see what's available
    print(f"  Display name: {model.display_name if hasattr(model, 'display_name') else 'N/A'}")
    print(f"  Supported methods: {model.supported_generation_methods if hasattr(model, 'supported_generation_methods') else 'N/A'}")
```

```python
import time
print("Waiting 30 seconds for API quota to reset...")
time.sleep(30)

sentence = "After years of hard work, your effort will take you"

# Call Gemini API
response = client.models.generate_content(
    model="models/gemini-2.5-pro", # Changed model to gemini-2.0-flash
    contents=sentence,
    config=types.GenerateContentConfig(
        max_output_tokens=1,      # Generate only the next token
        temperature=0,            # Deterministic output
        response_logprobs=True,   # Enable logprobs
        logprobs=20               # Request top 20 candidates (Max limit for Gemini)
    )
)

# --- Process Response ---

# 1. Access the logprobs result for the first generated step
# Gemini returns a list of 'top_candidates' for each generation step.
# Since max_output_tokens=1, we just want the first step (index 0).
logprobs_result = response.candidates[0].logprobs_result
top_candidates = logprobs_result.top_candidates[0].candidates

# 2. Convert to the dictionary format expected by the plotting code
# Format: {token_string: probability_value}
# We convert log_probability back to probability using np.exp()
probs = {c.token: np.exp(c.log_probability) for c in top_candidates}

# --- Visualization (Original Logic Preserved) ---

# Sort tokens by probability (highest first)
sorted_tokens = sorted(probs.items(), key=lambda x: x[1], reverse=True)

print("\nTop tokens with their probabilities:")
for token, prob in sorted_tokens[:10]:
    print(f"'{token}': {prob:.3f}")

# Unzip for plotting
tokens, probabilities = zip(*sorted_tokens)

# Create the plot
plt.figure(figsize=(12, 8))
plt.barh(range(len(tokens)), probabilities)
plt.yticks(range(len(tokens)), tokens)
plt.gca().invert_yaxis() # Highest probability on top
plt.ylabel("Token")
plt.xlabel("Probability")
plt.title(f"Distribution of next token Probability (Top {len(tokens)} tokens)")
plt.show()
```
